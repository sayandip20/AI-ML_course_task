```python
people = ('G1','G2','G3','G4','G5','G6','G7','G8')
segments = 4
data = [[3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039,7.1123587, 12.77792868, 3.44773477],
        [11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847],
        [3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433],
        [4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]]

percentages = (np.random.randint(5, 20, (len(people), segments)))
y_pos = np.arange(len(people))

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)

colors ='rgym'
patch_handles = []

left = np.zeros(len(people)) 
for i, d in enumerate(data):
    patch_handles.append(ax.barh(y_pos, d, 
      color=colors[i%len(colors)], align='center', 
      left=left))
    left += d

for j in range(len(patch_handles)):
    for i, patch in enumerate(patch_handles[j].get_children()):
        bl = patch.get_xy()
        x = 0.5*patch.get_width() + bl[0]
        y = 0.5*patch.get_height() + bl[1]
        ax.text(x,y, "%d%%" % (percentages[i,j]), ha='center')

ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.set_xlabel('Scores')
plt.show()
```

```python
#Ques 3: Write a Python program to add textures (black and white) to bars and wedges.
fig=plt.figure()
patterns = ['|','\\','/','+','-','.','*','X','o','O']
ax=fig.add_subplot(111)
for i in range(len(patterns)):
    ax.bar(i,3,color='white',edgecolor='black',hatch=patterns[i])

plt.show()
```

```python
# Now all other ques use this dataset..ok. These are Analysis questions..
import pandas as pd
df = pd.read_csv('company_sales_data.csv')
df
```

```python
#Ques 5: Read Total profit of all months and show it using a line plot..
plt.plot(df['month_number'],df['total_profit'],color='red')
plt.title('Company profit per month')
plt.xticks(np.arange(1,13))
plt.yticks(np.arange(100000,600000,100000))
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.show()
```

```python
'''Ques 6: Get Total profit of all months and show line plot with the following Style properties..
Generated line plot must include following Style properties: 
    Line Style dotted and Line-color should be red
    Show legend at the lower right location.
    X label name = Month Number
    Y label name = Sold units number
    Add a circle marker.
    Line marker color as read
    Line width should be 3'''
plt.plot(df['month_number'],df['total_profit'],'--r',marker='o',mfc='k',linewidth=2,label='Profit data of last year')
plt.title('Company Sales data of last year')
plt.xticks(np.arange(1,13))
plt.yticks(np.arange(100000,600000,100000))
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.legend()
plt.show()
```

```python
'''Ques 7: Read all product sales data and show it  using a multiline plot..
    Display the number of units sold per month for each product using multiline plots.'''
plt.plot(df['month_number'],df['facecream'],marker='o',label='Face Cream Sales Data',linewidth=2)
plt.plot(df['month_number'],df['facewash'],marker='o',label='Face Wash Sales Data',linewidth=2)
plt.plot(df['month_number'],df['toothpaste'],marker='o',label='Tooth Paste Sales Data',linewidth=2)
plt.plot(df['month_number'],df['bathingsoap'],marker='o',label='Bathing Soap Sales Data',linewidth=2)
plt.plot(df['month_number'],df['shampoo'],marker='o',label='Shampoo Sales Data',linewidth=2)
plt.plot(df['month_number'],df['moisturizer'],marker='o',label='Moisturizer Sales Data',linewidth=2)
plt.title('Sales data')
plt.xticks(np.arange(1,13))
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.legend()
plt.show()
```

```python
'''Ques 8: Read face cream and facewash product sales data and show it using the bar chart..
    The bar chart should display the number of units sold per month for each product. 
    Add a separate bar for each product in the same chart.'''
width=0.25
x_axis=np.arange(1,13)
plt.bar(x_axis - 0.15, df['facecream'], width, label='Face Cream Sales Data')
plt.bar(x_axis + 0.15, df['facewash'], width, label='Face Wash Sales Data')
plt.title('Facewash and facecream sales data')
plt.grid(linestyle='--')
plt.xticks(x_axis)
plt.yticks(np.arange(0,4000,500))
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.show()
```

```python
#Ques 13: Read all product sales data and show it using the stack plot..
monthList  = df ['month_number'].tolist()

faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingsoapSalesData   = df ['bathingsoap'].tolist()
shampooSalesData   = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(monthList, faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
plt.title('All product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
