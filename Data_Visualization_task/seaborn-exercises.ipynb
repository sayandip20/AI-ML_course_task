```python
'''Ques 1: dataset-- titanic
    Create the fare distribution through a histogram as shown below.'''
titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')
sns.displot(titanic['fare'],bins=30,kde=False,color='salmon')
plt.show()
```

```python
'''Ques 2: dataset-- titanic
    Write a python program to create a boxplot for age and class group of all passengers.'''
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
plt.show()
```

```python
'''Ques 3: dataset-- titanic 
    Create a bar chart for the number of male and female passengers.'''
sns.countplot(x='sex',data=titanic)
plt.show()
```

```python
'''Ques 4: dataset-- titanic
    Set up a grid to plot survival probability against several variables like class, sex, who and alone.
    Then Draw a seaborn pointplot onto each Axes'''
fig, axes = plt.subplots(1, 4, figsize=(8, 4))

variables = ['pclass', 'sex', 'who', 'alone']
titles = ['Class', 'Sex', 'Who', 'Alone']

for i, var in enumerate(variables):
    sns.pointplot(x=var, y='survived', data=titanic, ax=axes[i], color='violet')
    axes[i].set_yticks(np.arange(0.0,1.2,0.2))

for ax in axes[1:]:
    ax.set_ylabel('')
    ax.set_yticklabels([])
    
plt.tight_layout()
plt.show()
```

```python
'''Ques 5: dataset-- diamonds
    Generate the distribution of prices with respect to color for each cut quality.'''
diamond=sns.load_dataset('diamonds')
sns.barplot(x='cut',y='price',data=diamond,hue='color',hue_order=['E','I','J','H','F','G','D'])
plt.grid(False)
plt.legend(loc='upper right',ncol=4)
plt.show()
```

```python
'''Ques 6: dataset-- diamonds
    obtain the counts of diamonds of each cut quality through bar chart.'''
sns.countplot(x='cut',data=diamond,color='red')
plt.grid(False)
plt.show()
```

```python
'''Ques 7: dataset-- diamonds
Create a histogram of the frequency of diamonds in the dataset with their respective carat specifications on the x-axis.'''
sns.distplot(diamond['carat'],kde=False)
plt.grid(False)
plt.show()
```

```python
'''Ques 8: dataset-- tips
    Draw a nested violinplot and split the violins for easier comparison between day and total_bill.'''
tips=sns.load_dataset('tips')
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',palette=['lightblue','yellow'],split=True,inner='quartile')
plt.show()
```

```python
'''Ques 9: dataset-- titanic
    Draw a nested barplot to show survival for class and sex'''
sns.barplot(x='class',y='survived',data=titanic,hue='sex')
plt.show()
```

```python
'''Ques 11: dataset-- iris
    Create a scatter plot between sepal length and sepal width categorised by species.'''
iris=sns.load_dataset('iris')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
plt.grid(False)
plt.show()
```

```python
'''Ques 12: dataset-- iris
    Visualize the kernel density estimate of petal length for each species.'''
sns.FacetGrid(iris,hue='species').map(sns.kdeplot,'petal_length').add_legend()
plt.xticks(np.arange(1,9))
plt.grid(False)
plt.show()
```

```python
'''Ques 13: dataset-- iris
    Study the relationship between sepal length and sepal width for the iris dataset by creating a joint plot.'''
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
ax=plt.gca()
ax.grid(False)
plt.show()
```

```python
'''Ques 14: dataset-- iris
    Create a pairplot for the dataset to show relation between each pair of features based on different species.'''
sns.pairplot(iris,hue='species',height=2)
ax=plt.gca()
ax.grid(False)
ax.spines[['left','bottom']].set_color('black')
plt.show()
```

```python
'''Ques 15: dataset-- exercise
    Draw a pointplot to show pulse as a function of three categorical factors rest, walking and running.'''
exercise=sns.load_dataset('exercise')
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
custom_palette = sns.color_palette("Blues", n_colors=3)
sns.set_style('whitegrid')

df_no_fat = exercise[exercise['diet'] == 'no fat']

sns.pointplot(ax=axes[0], x='time', y='pulse', hue='kind', data=df_no_fat, palette=custom_palette)

df_no_fat = exercise[exercise['diet'] == 'low fat']

plot2 = sns.pointplot(ax=axes[1], x='time', y='pulse', hue='kind', data=df_no_fat, palette=custom_palette)

axes[0].set_title('diet = no fat')
axes[1].set_title('diet = low fat')

for ax in axes:
    ax.spines[['left','right','top']].set_visible(False)
    ax.set_yticks(range(90,150,10))

axes[0].get_legend().remove()
axes[1].set_ylabel('')
axes[1].set_yticklabels([])

plt.tight_layout()
plt.show()
```
